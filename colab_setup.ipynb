## Step 6: Verify Setup

Run this cell to verify everything is in place.

```python
import os

print("ðŸ“‚ Project Structure:\n")

def print_tree(path, prefix=""):
    items = sorted(os.listdir(path))
    for i, item in enumerate(items):
        full_path = os.path.join(path, item)
        is_last = i == len(items) - 1
        print(f"{prefix}{'â””â”€â”€ ' if is_last else 'â”œâ”€â”€ '}{item}")
        if os.path.isdir(full_path):
            print_tree(full_path, prefix + ('    ' if is_last else 'â”‚   '))

print_tree('.')

print("\n" + "="*50)
print("âœ… Setup complete! You can now run your experiments.")
print("="*50)
```

### Option B: Non-Interactive Mode with Custom Parameters

Configure your experiment parameters below and run:

```python
# ============================================================
# EXPERIMENT CONFIGURATION - Article-Aligned Settings
# Based on Wernicke 2005 paper recommendations
# ============================================================

import shutil
from google.colab import files

# Datasets to process (space-separated)
# Replaced roadNet-CA with Email-Enron (more interesting for motifs, faster)

DATASETS = "Wiki-Vote Amazon0302 CA-AstroPh"
# DATASETS = "roadNet-CA"
# DATASETS = "Wiki-Vote Amazon0302 CA-AstroPh Email-Enron"

# Motif size (k=3 for triads, aligned with article's significance analysis)
K_VALUES = "5"
# Sampling fraction (0.1 = 10% sampling, as per article)
Q_VALUES = "0.0001"
# Depth probability schedule ('fine' recommended per article Figure 3b)
SCHEDULE = "fine"
# Child selection strategy (bernoulli = paper default; balanced = footnote variance reduction)
CHILD_SELECTION = "bernoulli"
# Random seeds for variance estimation
SEEDS = "1 2 3"
# Memory optimization (recommended for k >= 4)
MEMORY_OPTIMIZED = True
# Baseline algorithm settings
BASELINE = "esa"
BASELINE_SAMPLES = 5000
ESA_PROBABILITY_CORRECTION = True  # Enable Equation 1 correction
# Significance analysis settings (edge-swap + direct for directed k=3 only)
SIGNIFICANCE_METHOD = "both"  # Runs only for directed k=3 datasets
SIGNIFICANCE_RANDOM_GRAPHS = 100  # Use 100 for speed; paper Table 2 uses 10,000
# Leave swaps-per-edge unset to use mfinder defaults (100 directed, 10 undirected)
SIGNIFICANCE_FULL_ENUMERATION = True  # Full enumeration for edge-swap random graphs
DIRECT_T = 100000  # Triples sampled for direct (BC) method
INCLUDE_LAMBDA_CORRECTION = True  # Include Î» correction term in BC formula
# Output directory
OUTPUT_DIR = "results_k3"

# ============================================================
# Build and run the command
# ============================================================

import multiprocessing
print(f"ðŸ–¥ï¸ Available CPU cores: {multiprocessing.cpu_count()}")

sig_enum_flag = (
    "--significance-full-enumeration"
    if SIGNIFICANCE_FULL_ENUMERATION
    else "--no-significance-full-enumeration"
)
cmd = f"""python -m src.experiments.run_all_datasets_parallel \
    --no-interactive \
    --datasets {DATASETS} \
    --k {K_VALUES} \
    --q {Q_VALUES} \
    --schedule {SCHEDULE} \
    --child-selection {CHILD_SELECTION} \
    --seed {SEEDS} \
    --baseline {BASELINE} \
    --baseline-samples {BASELINE_SAMPLES} \
    {'--esa-probability-correction' if ESA_PROBABILITY_CORRECTION else ''} \
    --significance-method {SIGNIFICANCE_METHOD} \
    --significance-random-graphs {SIGNIFICANCE_RANDOM_GRAPHS} \
    {sig_enum_flag} \
    --direct-T {DIRECT_T} \
    {'--include-lambda-correction' if INCLUDE_LAMBDA_CORRECTION else ''} \
    --output-dir {OUTPUT_DIR} \
    {'--memory-optimized' if MEMORY_OPTIMIZED else '--no-memory-optimized'}"""

print(f"\nðŸ“‹ Command to execute:")
print("=" * 70)
print(cmd)
print("=" * 70)
print("\nâ³ Starting experiment (this may take a while)...")
!{cmd}

# Create a zip file of the results
shutil.make_archive('results_colab', 'zip', OUTPUT_DIR)

print(f"ðŸ“¦ Created results_colab.zip from {OUTPUT_DIR}/")
print("ðŸ“¥ Downloading...")

# Download the zip file
files.download('results_colab.zip')
```
